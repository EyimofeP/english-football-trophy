Which Clubs have won Champions League?

```python
champs_df = pl[pl["UEFAChampions\'League"] >= 1][["Club", "UEFAChampions\'League"]].sort_values("UEFAChampions\'League", ascending =False)
champs_club = champs_df["Club"]
champs_number = champs_df["UEFAChampions\'League"]
```

```python
champs_df
```

Only 5 clubs in England have won the Champions League

```python
len(pl[pl["UEFAChampions\'League"] >= 1])
```

```python
sns.set_style("darkgrid")

plt.figure(figsize=[10,8])
plt.bar(champs_club, champs_number)

plt.title("English Champions League Winners")
plt.xlabel("Clubs")
plt.ylabel("Number of Trophies");
```

Which Clubs have the Premier League?

```python
league_df = pl[pl["League"] >= 1][["Club", "League"]].sort_values("League", ascending =False)
league_df
```

24 clubs in England have won the Premier League

```python
len(pl[pl["League"] >= 1])
```

```python
# Take Top 10 Winners
league_df = league_df.head(10)

# Increase Size of Graph
plt.figure(figsize=[20,8])

# Plot League against Club
plt.bar(x="Club", height="League", data=league_df)

plt.title("Top 10 Premier League Winners")
plt.xlabel("Clubs")
plt.ylabel("Number of Trophies");
```

Search for Clubs with "Town", "City", "United" in their names

```python
def club_finder(name, club):
    if name in club.lower().split():
        return True
    else:
        return False
```

Clubs with "Town"

```python
town = pl[pl["Club"].apply(lambda x : club_finder("town", x))]
town
```

```python
len(town)
```

Clubs with "City"

```python
city = pl[pl["Club"].apply(lambda x : club_finder("city", x))]
city
```

```python
len(city)
```

Clubs with "United"

```python
united = pl[pl["Club"].apply(lambda x : club_finder("united", x))]
united
```

```python
len(united)
```
